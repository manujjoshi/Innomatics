```python
X = data.drop('price',axis=1)
y = data['price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

### `KNN from scratch`  
- [Distance Metric](https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.DistanceMetric.html)

## `KNN from scratch`

![](https://www.saedsayad.com/images/KNN_similarity.png)
The above three distance measures are only valid for continuous variables. In the case of categorical variables you must use the Hamming distance, which is a measure of the number of instances in which corresponding symbols are different in two strings of equal length. 		
![](https://www.saedsayad.com/images/KNN_hamming.png)
but as here variables are of mixed types I will use **Gower distance**

## `KNN sklearn`

```python
from sklearn.neighbors import KNeighborsRegressor
knr = KNeighborsRegressor()   #by-default n_neighbors=5
knr.fit(X_train,y_train)
y_pred = knr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
print(mean_squared_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```
