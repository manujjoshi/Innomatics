```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_excel('aspiring_minds_employability_outcomes_2015.xlsx')
data.head()
```

```python
data.drop(['Unnamed: 0','ID'],axis=1,inplace=True)
```

```python
data.shape
```

```python
data.info()
```

#### `This is not true, data contain null vales as -1`

```python
data.isin([-1, 'NaN']).sum() #finding number of null value
```

#### `THis is not true, data contain null vales as 0`

```python
data.isin([0,'NaN']).sum() #finding number of null value
```

```python
data.replace(-1,np.NaN,inplace=True)
```

```python
data.replace(0,np.NaN,inplace=True)
```

#### `These are the true null values`

```python
data.isnull().sum()
```

#### `Missing data in percentage`

```python
(data.isnull().sum()/data.shape[0])*100
```

#### `Due to huge amount of column data is missing, I am dropping that columns`

```python
data.drop(['ElectronicsAndSemicon','ComputerScience','MechanicalEngg','ElectricalEngg','TelecomEngg','CivilEngg','CollegeCityTier'],axis=1,inplace=True)
```

```python
data.skew()
```

```python
data.kurtosis()
```

```python
cont_columns = data.select_dtypes(['int64','float64']).columns.to_list()
cat_columns =  data.select_dtypes(['object']).columns.to_list()
```

```python
cat_columns.remove('DOL')  # as DOL is datetype
```

```python
cont_columns
```

```python
cont_columns.remove('CollegeTier')
cont_columns.remove('CollegeID')
cont_columns.remove('CollegeCityID')
cont_columns.remove('GraduationYear')
```

```python
cat_columns
```

## `Univariate Analysis ----------------------------------------------------`
### `Numerical columns`

```python
for i in cont_columns:
    sns.distplot(data[i])
    plt.show()
```

```python
for i in cont_columns:
    sns.boxplot(data[i])
    plt.show()
```

### `Categorical Columns`

```python
specialization_freq = data['Designation'].value_counts()[:20]
specialization_freq.plot(kind='bar', figsize=(15,5))
```

```python
sns.countplot(data['Gender'])
```
